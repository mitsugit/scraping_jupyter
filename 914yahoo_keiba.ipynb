```python
from urllib import request
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url_list = ["https://keiba.yahoo.co.jp/odds/tfw/2006040101/","https://keiba.yahoo.co.jp/odds/tfw/2006040102/"]
```

```python
df_list = []
race_ID_list = []
for url in url_list:
    response = request.urlopen(url)
    soup = BeautifulSoup(response)
    open_date=soup.find(id="raceTitDay").text.split('|')[0].replace(" ","")
    kaisai = soup.find(id="raceTitDay").text.split("|")[1]
    race_No=soup.find(id="raceNo").text
    print(race_No)
    df_list.append(pd.read_html(url)[2:4])
    race_ID_list.append(open_date+"_"+kaisai+"_"+race_No)
```

```python
race_ID_list[0]
```

```python
df_list_mod = []
for num,i in enumerate(df_list):
    x =pd.concat([i[0],i[1]], ignore_index=True,axis=0)
    x = x.assign(race_ID=race_ID_list[num],開催日=open_date,開催場所=kaisai,レース番号=race_No)
    df_list_mod.append(x)
```

```python
df_list_mod
```

```python
df = df_list_mod[0]
for i in df_list_mod:
    df = pd.concat([df,i], ignore_index=True,axis=0)
```
